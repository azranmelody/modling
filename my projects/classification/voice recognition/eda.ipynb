```python
df = pd.read_csv('dataset/voice.csv')
```

```python
labels = {'male' : 1 , 'female' : 2}
df['label'] = df['label'].map(labels)
```

```python
df['IQR_BIN'] = pd.qcut(df['IQR'] , 4 , range(4))
```

```python
df
```

```python
plt.figure(figsize=(2,10))
sns.heatmap(pd.DataFrame(abs(df.corr()['label'])) ,annot=True , vmin=0.4)
```

```python
sns.boxplot(data=df ,x=df['IQR_BIN'], y=df['meanfun'], hue ='label')
```

```python
df_p = df.pivot_table(index='IQR_BIN', columns='label',values='meanfun', aggfunc='count')
```

```python
df_p
```

```python
fig, ax = plt.subplots()  
```

```python
ax = df_p.plot( kind='bar', stacked=True)
```

```python
df = df_p
ax = df.plot.bar(stacked = True);
#print(df)
for rowNum,row in df.iterrows():
    xpos = 0
    for val in row:
        xpos += val
        ax.text(xpos , rowNum-0.05, str(val), color='black')
#     xpos = 0
# display(ax)
```
